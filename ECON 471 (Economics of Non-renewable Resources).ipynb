Pareto Optimal Allocation of resources is achieved at the point where Marginal Rate of Substitution (MRS) is equal to Marginal Rate of Transformation (MRT). 

$$ MRS = \frac{MU_x}{MU_y}, MRT = \frac{MC_x}{MC_y}$$ 

If $MRS > MRT$, then relative utility from one more X is higher than is MC. So efficiency requires that more X be produced. When $MRS = MRT$, we achieve both production and consumption (allocative) efficiency. 

Monopolies want to maximize profits ($\pi$) at $P(Q)*Q - C(Q)$. Profit maximization occurs at $MR(Q)=MC(Q)$. 

Welfare loss (DWL) under monopoly is triangle ABC. Notice that $P_m$ > MR = MC.

## Topic 3 - Externalities

- Externality arises when a person engages in an activity (production or consumption) that influences the well-being of a bystander and yet neither pays nor receives any compensation for that effect.Externalities can be positive or negative depending on the action of one party imposes a cost or benefit to another. Example: Pollution is an externality (negative) because the polluter does not pay for the damages it causes to the environment. 

Example: Two firms; Paper Factory (P) and Apple Producer (A). Paper Factory discharges wastes during production which harms Apple producer's production of apples. We have the production functions Apple = $Q_1$ = A(x,z) and Paper = $Q_2$ = P(y), where x and y are inputs of firms A and P, and z is the waste produced by the paper prodction depending on the level of y use. 
Pollution from paper, z, negatively affects A's production s.t. $A_z = \frac{dA}{dz} < 0$. Also as more y is used, the more pollution is produced; $\frac{dz}{dy} > 0$. 

In a free market, each firm will maximize it's own profits independently of other firms. 

(1) Max $\pi_A = P_AQ_A-P_xX$

(2) Max $\pi_P = P_pQ_p-P_yY$

Each firm then chooses its factor x or y which satisfies Marginal Benefit (MB) = 0. 

In a socially optimal solution, efficiency requires that both firms maximize joint profits: 

(3) Max $\pi = P_AQ_A-P_xX + P_PQ_P-P_yY$ 

Efficiency requires MB(y) = MD(y). 

By taking the derivative, we can see that z harms the production of apple, which makes $\frac{dA}{dz} < 0, \frac{dz}{dy} > 0$ because the production of paper will increase the production of pollution. 
$\rightarrow \frac{dA}{dz}\frac{dz}{dy} < 0$ means that the dollar value of marginal damage caused by one unit of y by the paper factory. 

Taxes or Subsidies

- Govn't can solve problem caused by negative externalities by imposing a tax, t = Marginal External Cost (MEC) or MD at $Q_{opt}$. 
- Tax on producer will increase MC of production of the factory by the tax amount and shits MC of production (after tax) to the left by tax amount. 
- Tax induces producer to reduce equilibrium quantity to socially desirable quantity eliminating DWL. 
- Optimal Solution is t = MB = MD. 

Transferring Emissions Permits

- Polluters can trade emission permits. 
- High cost firms will purchase permits (and pollute more) from low cost firms (who will control more pollutions)
- Firms whose MC of pollution > Price of permits will buy the permits until $MC = p$. 
- Firms whose MC of pollution < Price of permits, will sell the permits until $MC = p$. 
- Trade stops at $p = MC$. 

Emission Taxes vs. Emission Standard

- Emission Standard sets a legal limit on emissions at optimal solution E*.
- This can be enforced by high enough monetary penalities in the case of violations. 

- Emission Tax or Fees: 
    - Polluter's MB > tax, then polluter will emit more and pay the tax. 
    - Polluters alwasy pollute at t = MB. 
    - Under a free market without govn't regulation, polluters pollute where MB = 0. 
    
If regulator doesn't know MB of polluter (or MC of pollution control), they can impose a tax, t, or $E_{std}$, at the level based on expected efficient solution, where govn't expects MEC = expected MB. 

Case 1: MB is flatter than MEC. 


Case 2: MB is steeper than MEC. 


Coarse Theorem: When parties can bargain without transaction costs, the resulting outcome will be efficient, regardless of how property rights are specified. 
- Bargaining is costly as it requres clearly defined property rights. 
- Bargaining is difficult when multiple parties are involved. 

Example: MB = 12-y, and MD = 2y. Market solution, $y_m$ = 12, whereas equilibrium $y* = 4$. 

If polluting firm has the right the pollute, it will pollute at $y_m$. The victim has MD > MB, and is willing to pay to reduce pollution to $y*$. 

If the victim has a right to clean environment, then the polluting firm must produce $y = 0$. Since MB > MD, the polluting firm is willing to bribe the victim until $y=y*$.

## Topic 5 - Non-renewable Resource: Theory of Depletion

Recall: Non-renewable resources include oil, gas, coal, uranium, zinc, etc. 

Under a competitive market for a reproducible good, profit maximization condition is p = MR = MC. 

Profit maximization for a non-renewable good is p = MR = MC+opportunity cost of depletion. 

**When the discount rate is positive, the rent is positive and rises in nominal value as depletion occurs. If rent doesn't rise in value, the miner would want to extract it as fast as the technology would allow and put the proceeds in a savings account that earns higher interest than the resource rent (ie. $\frac{P_1-P_0}{P_0} < r$) If the value of the resource is growing at a rate faster than the return on alternative investments, there is no incentives to extract at all (ie. $\frac{P_1-P_0}{P_0} > r $)** 

Competitive Firm: Constant Price 
- Want to maximize profits $\pi$ such that: $L = \sum_{t=0}^T\frac{pq_t - C(q_t)}{(1+r)^t} + \lambda(S-\sum_{n=1}^N\sum_{t=0}^T q_t)$. 
- Differentiate $\frac{dL}{dq_t}$ and $\frac{dL}{dq_{t+1}}$ and $\frac{dL}{q\lambda}$
- Then set $\frac{dL}{dq_t}$ = $\frac{dL}{dq_{t+1}}$ and solve subject to the constraint $\frac{dL}{q\lambda}$
- Undiscounted Hotelling Rent is $\lambda S = \frac{p-MC(q_{t+1})-(p-MC(q_t))}{p-MC(q_t)} = r$

Competitive Firm: Non-Constant Price
- Similarly set up Lagrangian and get $\lambda S = \frac{p_{t+1}-MC(q_{t+1})-(p_t-MC(q_t))}{p_t-MC(q_t)} = r$

Efficiency extraction of resource requires that the present value of the rent must be the same in each period (and equal to $\lambda$. 

Example: Consider two periods, then we get $\frac{p(Q_0)-MC(Q_0)}{(1+r)^0} = \frac{p(Q_1)-MC(Q_1)}{(1+r)^1}$

Price and Quantity Path Analysis 
- $P_T$ is the backstop or choking price
- Backstop price is the maximum price level that a non-renewable can reach. 
- Effect of Change in Extraction Costs: 
    - If extraction costs fall, then price path falls and resource is depleted sooner. 
    


- Effect of Change in Discount Rate: 
    - If discount rate increases, then people value the present more $\rightarrow$ less savings and more spending $\rightarrow$ resource is extracted and depleted sooner. 
    

- Effect of Change in Choke Price: 
    - An increase in extraction technology makes extraction cheaper, thus choking price falls. Resource will be depleted sooner.

## Topic 7 - Energy Economics
- Prisoner's Dilemma 
    - Nash Equilibrium is when each player does what is in their best interest, given the opponent's action. 
    - Dominant Strategy: A strategy that one player chooses every time. 
    - When a prisoner's dilemma game is played for infinite times, the benefits from co-operation is much higher than the one time benefit one gets from defecting. 
    - The player that defects will benefit immediately, but will be punished forever as long as the game continues and may not be beneficial relative to playing the game forever. 

- Dominant Firm Model: Large firms acts as dominant firms, setting prices that maximizes its own profits. 
- Fringe firms take prices that is given and produces its output consistent with that price 
- Dominant Firm produces $Q_d$ where $MR_D = MC_D$. 
- Fringe Firm takes this P* and sells $Q_F$ where $P* = MC_F$

- Peak Load Pricing: Demand for energy at particular peak times/seasons
    - Example: Hotel Prices during vacation times
- Block Pricing: Quantity Based Price Discrimination
    - Example: BC Hydro charging 6.9 cents/kwh for the first 1350kwh and 10.34 cents/kwh for the rest of the billing period. 
    - Principle behind Block Pricing is to encourage conservation of resources. 

- Natural Monopolies arises due to economies of scale, when one firm can meet the demand at lower average costs (AC) than two or more firms. 
    - MC < AC for all levels of production such that AC decreases as levels of production increase 
    - Economies of Scale: When a firm increases production by 1%, then costs increase by less than 1%
    

- An unregulated monopoly would set maximize profits by producing $Q_m$ by setting MR = MC and charge price $P_m$ at the demand curve. This results in a DWL. There are profits. 
- DWL can be eliminated by having regulation to make monopolies set price at $P = MC$ where MC intersects the demand curve. This comes at a cost of a loss to the natural monopoly in the area $Q_c*(AC-P_c)$


- Average Cost Pricing (AC Pricing) is where regulation sets $P = AC$, where AC intersects the demand curve. This solution results in a lowered DWL from monopoly $P_m$, but profits = 0. 


- Ramsay Pricing or Inverse Elasticity Rule, raises individual prices above MC according to each service's price elasticity of demand. 
- Inverse Elasticity Rule: Mark ups above MC are lower for services with more than elastic demand and conversely greater for services with more inelastic demand. 
- Two Tiered Pricing System: Monopoly is allowed to charge a higher price (monopoly price) to groups of users and lower price (equal to MC) to marginal users. 
- Two part tariff: Consists of fixed payment (regardless of consumption) plus a price, p, per unit of consumption. 
- Loeb-Magat (LM) Model
    - Regulator allows monopolist to choose its price and subsidizdes it by the CS at that price
    - P = $P_m$, then the monopolist gets Revenue - Variable Cost + Consumer Surplus (at this price). 
    
    - Problem: All surplus accrues to the monopolist leaving no surplus at all for consumers, while optimal solution is achieved.
